```python
#############################################
# COMPLETE TRAINING & EVALUATION PIPELINE   #
# FOR DEBLURRING (60 Epochs)                #
# (Multi-Scale U-Net + CBAM + Self-Attn)      #
#############################################

import os
import random
import math
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

import torchvision.transforms as transforms
import torchvision.transforms.functional as TF
import torchvision.models as models

from skimage.metrics import structural_similarity as ssim
from tqdm import tqdm

# If using Google Colab, mount the drive
try:
    from google.colab import drive
    drive.mount('/content/drive', force_remount=False)
except Exception as e:
    print("Not running in Colab? Skipping drive.mount().", e)

#############################################
# 1. PATH CONFIGURATION
#############################################
dataset_path = "/content/gopro_deblur"
blur_dir  = os.path.join(dataset_path, "blur/images")
sharp_dir = os.path.join(dataset_path, "sharp/images")

# Define output directories in your Google Drive
base_output_dir = "/content/drive/MyDrive/deblur_selfattn_60"
os.makedirs(base_output_dir, exist_ok=True)
eval_output_dir = os.path.join(base_output_dir, "outputs_of_60epoches_deblurred_testing_images")
os.makedirs(eval_output_dir, exist_ok=True)

#############################################
# 2. DATA SPLIT & DATASET DEFINITION
#############################################
all_files = sorted(os.listdir(blur_dir))
# Ensure each file exists in both folders
all_files = [f for f in all_files if os.path.exists(os.path.join(sharp_dir, f))]
random.shuffle(all_files)
n_total = len(all_files)
n_train = int(0.8 * n_total)
n_val   = int(0.1 * n_total)
train_list = all_files[:n_train]
val_list   = all_files[n_train:n_train+n_val]
test_list  = all_files[n_train+n_val:]
print(f"Total images: {n_total} | Train: {len(train_list)} | Val: {len(val_list)} | Test: {len(test_list)}")

class GoProDataset(Dataset):
    def __init__(self, blur_dir, sharp_dir, file_list, training=True):
        self.blur_dir = blur_dir
        self.sharp_dir = sharp_dir
        self.file_list = file_list
        self.training = training  # no extra augmentation as requested

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, idx):
        fname = self.file_list[idx]
        blur_path  = os.path.join(self.blur_dir, fname)
        sharp_path = os.path.join(self.sharp_dir, fname)
        blur_img  = Image.open(blur_path).convert('RGB')
        sharp_img = Image.open(sharp_path).convert('RGB')
        crop_size = 256
        w, h = blur_img.size
        if w >= crop_size and h >= crop_size:
            left = (w - crop_size) // 2
            top = (h - crop_size) // 2
            blur_img = blur_img.crop((left, top, left + crop_size, top + crop_size))
            sharp_img = sharp_img.crop((left, top, left + crop_size, top + crop_size))
        else:
            blur_img = blur_img.resize((crop_size, crop_size))
            sharp_img = sharp_img.resize((crop_size, crop_size))
        blur_t = TF.to_tensor(blur_img)
        sharp_t = TF.to_tensor(sharp_img)
        return blur_t, sharp_t

train_ds = GoProDataset(blur_dir, sharp_dir, train_list, training=True)
val_ds   = GoProDataset(blur_dir, sharp_dir, val_list, training=False)
test_ds  = GoProDataset(blur_dir, sharp_dir, test_list, training=False)

train_loader = DataLoader(train_ds, batch_size=4, shuffle=True, num_workers=2)
val_loader   = DataLoader(val_ds, batch_size=1, shuffle=False, num_workers=2)
test_loader  = DataLoader(test_ds, batch_size=1, shuffle=False, num_workers=2)

#############################################
# 3. MODEL DEFINITION: Multi-Scale U-Net with CBAM and Self-Attention
#############################################
# Self-Attention Block
class SelfAttnBlock(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.query = nn.Conv2d(channels, channels, 1)
        self.key   = nn.Conv2d(channels, channels, 1)
        self.value = nn.Conv2d(channels, channels, 1)
        self.gamma = nn.Parameter(torch.zeros(1))
        self.softmax = nn.Softmax(dim=-1)
    def forward(self, x):
        B, C, H, W = x.shape
        q = self.query(x).view(B, C, -1)
        k = self.key(x).view(B, C, -1)
        v = self.value(x).view(B, C, -1)
        attn = torch.bmm(q.permute(0, 2, 1), k)
        attn = self.softmax(attn / (C ** 0.5))
        out = torch.bmm(attn, v.permute(0, 2, 1))
        out = out.permute(0, 2, 1).view(B, C, H, W)
        return x + self.gamma * out

# CBAM: Channel and Spatial Attention
class ChannelAttention(nn.Module):
    def __init__(self, in_planes, ratio=16):
        super().__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)
        self.fc1 = nn.Conv2d(in_planes, in_planes // ratio, 1, bias=False)
        self.relu = nn.ReLU()
        self.fc2 = nn.Conv2d(in_planes // ratio, in_planes, 1, bias=False)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        avg_out = self.fc2(self.relu(self.fc1(self.avg_pool(x))))
        max_out = self.fc2(self.relu(self.fc1(self.max_pool(x))))
        return self.sigmoid(avg_out + max_out) * x

class SpatialAttention(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Conv2d(2, 1, 7, padding=3, bias=False)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        avg_out = torch.mean(x, dim=1, keepdim=True)
        cat = torch.cat([max_out, avg_out], dim=1)
        return self.sigmoid(self.conv(cat)) * x

class CBAMBlock(nn.Module):
    def __init__(self, ch):
        super().__init__()
        self.ca = ChannelAttention(ch)
        self.sa = SpatialAttention()
    def forward(self, x):
        x = self.ca(x)
        x = self.sa(x)
        return x

# Double Convolution Block
class DoubleConv(nn.Module):
    def __init__(self, in_c, out_c, dilate=False):
        super().__init__()
        d = 2 if dilate else 1
        pad = d
        self.block = nn.Sequential(
            nn.Conv2d(in_c, out_c, 3, padding=pad, dilation=d),
            nn.BatchNorm2d(out_c),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_c, out_c, 3, padding=pad, dilation=d),
            nn.BatchNorm2d(out_c),
            nn.ReLU(inplace=True)
        )
    def forward(self, x):
        return self.block(x)

class MultiScaleAttnUNet(nn.Module):
    def __init__(self, base=96):
        super().__init__()
        # Encoder
        self.enc1 = nn.Sequential(DoubleConv(3, base), CBAMBlock(base))
        self.enc2 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(base, base * 2), CBAMBlock(base * 2))
        self.enc3 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(base * 2, base * 4), CBAMBlock(base * 4))
        self.enc4 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(base * 4, base * 8), CBAMBlock(base * 8))
        # Bottleneck with dilated convolution and self-attention
        self.bottleneck = DoubleConv(base * 8, base * 8, dilate=True)
        self.self_attn = SelfAttnBlock(base * 8)
        # Decoder
        self.up3 = nn.ConvTranspose2d(base * 8, base * 4, 2, 2)
        self.dec3 = nn.Sequential(DoubleConv(base * 8, base * 4), CBAMBlock(base * 4))
        self.up2 = nn.ConvTranspose2d(base * 4, base * 2, 2, 2)
        self.dec2 = nn.Sequential(DoubleConv(base * 4, base * 2), CBAMBlock(base * 2))
        self.up1 = nn.ConvTranspose2d(base * 2, base, 2, 2)
        self.dec1 = nn.Sequential(DoubleConv(base * 2, base), CBAMBlock(base))
        # Multi-scale outputs (only final output is used for loss)
        self.out_final = nn.Conv2d(base, 3, 1)
        self.out_half  = nn.Conv2d(base * 2, 3, 1)
        self.out_qter  = nn.Conv2d(base * 4, 3, 1)
    def forward(self, x):
        e1 = self.enc1(x)
        e2 = self.enc2(e1)
        e3 = self.enc3(e2)
        e4 = self.enc4(e3)
        b = self.bottleneck(e4)
        b = self.self_attn(b)
        d3 = self.up3(b)
        if d3.size() != e3.size():
            diffY = e3.size(2) - d3.size(2)
            diffX = e3.size(3) - d3.size(3)
            d3 = F.pad(d3, (0, diffX, 0, diffY))
        d3 = self.dec3(torch.cat([d3, e3], dim=1))
        out_q = torch.clamp(self.out_qter(d3), 0, 1)
        d2 = self.up2(d3)
        if d2.size() != e2.size():
            diffY = e2.size(2) - d2.size(2)
            diffX = e2.size(3) - d2.size(3)
            d2 = F.pad(d2, (0, diffX, 0, diffY))
        d2 = self.dec2(torch.cat([d2, e2], dim=1))
        out_h = torch.clamp(self.out_half(d2), 0, 1)
        d1 = self.up1(d2)
        if d1.size() != e1.size():
            diffY = e1.size(2) - d1.size(2)
            diffX = e1.size(3) - d1.size(3)
            d1 = F.pad(d1, (0, diffX, 0, diffY))
        d1 = self.dec1(torch.cat([d1, e1], dim=1))
        out_f = torch.clamp(self.out_final(d1), 0, 1)
        return out_f, out_h, out_q

#############################################
# 4. LOSS FUNCTIONS & METRICS
#############################################
# Perceptual Loss (using VGG16 features up to relu2_2)
vgg16 = models.vgg16(weights=models.VGG16_Weights.IMAGENET1K_V1).features[:16].eval().to(torch.device("cuda" if torch.cuda.is_available() else "cpu"))
for p in vgg16.parameters():
    p.requires_grad = False

def perceptual_loss(pred, target):
    norm = transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    pn = norm(pred)
    tn = norm(target)
    fp = vgg16(pn)
    ft = vgg16(tn)
    return F.mse_loss(fp, ft)

def fourier_loss(pred, target):
    pfft = torch.fft.fftn(pred, dim=(-2, -1))
    tfft = torch.fft.fftn(target, dim=(-2, -1))
    return F.l1_loss(torch.abs(pfft), torch.abs(tfft))

critL1 = nn.L1Loss()
w_l1   = 1.0
w_perc = 0.1
w_four = 0.08

def calc_psnr(x, y):
    mse = F.mse_loss(x, y).item()
    if mse < 1e-12:
        return 100.0
    return 10. * math.log10(1.0 / mse)

def calc_ssim(x, y, win_size=7):
    # Convert tensor images [0,1] to uint8 for SSIM calculation.
    x_np = (x.squeeze(0).permute(1,2,0).cpu().numpy() * 255).astype(np.uint8)
    y_np = (y.squeeze(0).permute(1,2,0).cpu().numpy() * 255).astype(np.uint8)
    return ssim(x_np, y_np, data_range=255, channel_axis=2, win_size=win_size)

def upsample_to(pred, factor):
    if factor > 1:
        return F.interpolate(pred, scale_factor=factor, mode='bilinear', align_corners=False)
    return pred

#############################################
# 5. INITIALIZE MODEL, OPTIMIZER & SCHEDULER
#############################################
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MultiScaleAttnUNet(base=96).to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-4)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=60, eta_min=1e-6)
num_epochs = 60

#############################################
# 6. TRAINING LOOP (60 Epochs)
#############################################
train_losses = []
val_losses = []
best_val_loss = float('inf')

print("Starting Training...")
for epoch in range(1, num_epochs+1):
    model.train()
    running_loss = 0.0
    pbar_train = tqdm(train_loader, desc=f"Epoch {epoch}/{num_epochs} [Train]", leave=False)
    for blur_b, sharp_b in pbar_train:
        blur_b, sharp_b = blur_b.to(device), sharp_b.to(device)
        out_f, out_h, out_q = model(blur_b)
        up_h = upsample_to(out_h, 2)
        up_q = upsample_to(out_q, 4)
        lf = critL1(out_f, sharp_b)
        lh = critL1(up_h, sharp_b)
        lq = critL1(up_q, sharp_b)
        l_spatial = lf + 0.5 * lh + 0.25 * lq
        p_loss = perceptual_loss(out_f, sharp_b)
        f_loss = fourier_loss(out_f, sharp_b)
        loss = w_l1 * l_spatial + w_perc * p_loss + w_four * f_loss
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        pbar_train.set_postfix(loss=f"{loss.item():.4f}")
    train_epoch_loss = running_loss / len(train_loader)
    train_losses.append(train_epoch_loss)

    # Validation loop
    model.eval()
    val_sum = 0.0
    pbar_val = tqdm(val_loader, desc=f"Epoch {epoch}/{num_epochs} [Val]", leave=False)
    with torch.no_grad():
        for b_val, s_val in pbar_val:
            b_val, s_val = b_val.to(device), s_val.to(device)
            of, oh, oq = model(b_val)
            uh = upsample_to(oh, 2)
            uq = upsample_to(oq, 4)
            lf_val = critL1(of, s_val)
            lh_val = critL1(uh, s_val)
            lq_val = critL1(uq, s_val)
            l_sp = lf_val + 0.5 * lh_val + 0.25 * lq_val
            p_loss_val = perceptual_loss(of, s_val)
            f_loss_val = fourier_loss(of, s_val)
            val_loss = w_l1 * l_sp + w_perc * p_loss_val + w_four * f_loss_val
            val_sum += val_loss.item()
            pbar_val.set_postfix(loss=f"{val_loss.item():.4f}")
    epoch_val_loss = val_sum / len(val_loader)
    val_losses.append(epoch_val_loss)

    scheduler.step()

    # Save best model based on validation loss
    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        torch.save(model.state_dict(), os.path.join(base_output_dir, "best_model.pth"))

    print(f"Epoch {epoch}/{num_epochs} - Train Loss: {train_epoch_loss:.4f}, Val Loss: {epoch_val_loss:.4f}")

# Save and plot loss curve
plt.figure(figsize=(7,5))
plt.plot(range(1, num_epochs+1), train_losses, label='Train Loss')
plt.plot(range(1, num_epochs+1), val_losses, label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training & Validation Loss Curve')
plt.legend()
plt.grid(True)
loss_curve_path = os.path.join(base_output_dir, "loss_curve.png")
plt.savefig(loss_curve_path)
plt.show()
print("Loss curve saved to", loss_curve_path)

#############################################
# 7. TESTING & EVALUATION
#############################################
best_model_path = os.path.join(base_output_dir, "best_model.pth")
if os.path.exists(best_model_path):
    model.load_state_dict(torch.load(best_model_path, map_location=device))
    model.eval()
else:
    print("Best model weights not found. Exiting evaluation.")
    exit()

psnr_values = []
ssim_values = []
csv_data = []  # to store (filename, PSNR, SSIM)

print("Starting Evaluation on Test Set...")
with torch.no_grad():
    for fname in tqdm(test_list, desc="[Test Evaluation]"):
        blur_path = os.path.join(blur_dir, fname)
        sharp_path = os.path.join(sharp_dir, fname)
        if not os.path.exists(sharp_path):
            continue
        blur_img = Image.open(blur_path).convert('RGB')
        sharp_img = Image.open(sharp_path).convert('RGB')
        crop_size = 256
        w, h = blur_img.size
        if w >= crop_size and h >= crop_size:
            left = (w - crop_size) // 2
            top = (h - crop_size) // 2
            blur_img = blur_img.crop((left, top, left + crop_size, top + crop_size))
            sharp_img = sharp_img.crop((left, top, left + crop_size, top + crop_size))
        else:
            blur_img = blur_img.resize((crop_size, crop_size))
            sharp_img = sharp_img.resize((crop_size, crop_size))

        blur_tensor = TF.to_tensor(blur_img).unsqueeze(0).to(device)
        sharp_tensor = TF.to_tensor(sharp_img).unsqueeze(0).to(device)

        out_f, out_h, out_q = model(blur_tensor)
        deblur_tensor = torch.clamp(out_f, 0, 1)

        psnr_val = calc_psnr(deblur_tensor, sharp_tensor)
        try:
            ssim_val = calc_ssim(deblur_tensor, sharp_tensor, win_size=7)
        except Exception as e:
            print(f"SSIM error for {fname}: {e}")
            ssim_val = 0.0
        psnr_values.append(psnr_val)
        ssim_values.append(ssim_val)
        csv_data.append((fname, psnr_val, ssim_val))

        # Create side-by-side image: [Blur | Deblurred | Sharp]
        deblur_img = TF.to_pil_image(deblur_tensor.cpu().squeeze(0))
        combined_width = blur_img.width + deblur_img.width + sharp_img.width
        combined_img = Image.new("RGB", (combined_width, crop_size))
        combined_img.paste(blur_img, (0, 0))
        combined_img.paste(deblur_img, (blur_img.width, 0))
        combined_img.paste(sharp_img, (blur_img.width + deblur_img.width, 0))
        combined_save_path = os.path.join(eval_output_dir, fname)
        combined_img.save(combined_save_path, format="PNG", compress_level=6)

# Save CSV file with test metrics
csv_path = os.path.join(eval_output_dir, "results_metrics.csv")
with open(csv_path, "w") as f:
    f.write("Filename,PSNR,SSIM\n")
    for row in csv_data:
        f.write(f"{row[0]},{row[1]:.4f},{row[2]:.6f}\n")
print("CSV file with metrics saved to", csv_path)

psnr_arr = np.array(psnr_values)
ssim_arr = np.array(ssim_values)
avg_psnr = np.mean(psnr_arr)
avg_ssim = np.mean(ssim_arr)
std_psnr = np.std(psnr_arr)
std_ssim = np.std(ssim_arr)
print(f"Test PSNR: {avg_psnr:.2f} dB, Test SSIM: {avg_ssim:.4f}")
print(f"PSNR: Mean = {avg_psnr:.2f}, Std = {std_psnr:.2f}")
print(f"SSIM: Mean = {avg_ssim:.4f}, Std = {std_ssim:.4f}")

#############################################
# 8. PLOTTING HISTOGRAMS OF TEST METRICS
#############################################
plt.figure(figsize=(6,4))
plt.hist(psnr_values, bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Test PSNR")
plt.xlabel("PSNR (dB)")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.75)
hist_psnr_path = os.path.join(eval_output_dir, "histogram_psnr.png")
plt.savefig(hist_psnr_path)
plt.close()

plt.figure(figsize=(6,4))
plt.hist(ssim_values, bins=20, color='lightgreen', edgecolor='black')
plt.title("Histogram of Test SSIM")
plt.xlabel("SSIM")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.75)
hist_ssim_path = os.path.join(eval_output_dir, "histogram_ssim.png")
plt.savefig(hist_ssim_path)
plt.close()
print("Histograms saved.")

#############################################
# 9. MODEL SUMMARY & TRAINING LOG REPORT
#############################################
model_summary_str = ""
try:
    from torchinfo import summary as torch_summary
    summary_info = torch_summary(model, input_size=(1, 3, 256, 256), verbose=0)
    model_summary_str = str(summary_info)
except Exception as e:
    total_params = sum(p.numel() for p in model.parameters())
    model_summary_str = f"Model has {total_params} parameters.\n" + str(model)

summary_txt_path = os.path.join(eval_output_dir, "model_summary.txt")
with open(summary_txt_path, "w") as f:
    f.write("### Model Architecture Summary ###\n")
    f.write(model_summary_str + "\n\n")
    f.write("### Final Evaluation Metrics ###\n")
    f.write(f"Average PSNR: {avg_psnr:.4f} dB\n")
    f.write(f"Average SSIM: {avg_ssim:.6f}\n")
    f.write(f"PSNR: Mean = {avg_psnr:.4f}, Std = {std_psnr:.4f}\n")
    f.write(f"SSIM: Mean = {avg_ssim:.6f}, Std = {std_ssim:.6f}\n\n")
    f.write("### Training and Validation Loss per Epoch ###\n")
    f.write("Epoch,TrainingLoss,ValidationLoss\n")
    for epoch in range(num_epochs):
        f.write(f"{epoch+1},{train_losses[epoch]:.6f},{val_losses[epoch]:.6f}\n")
print("Model summary and training log saved to", summary_txt_path)

print("All outputs saved in:", eval_output_dir)
```
